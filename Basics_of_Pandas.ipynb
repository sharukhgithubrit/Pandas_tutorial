```python
import numpy as np
import pandas as pd
```

```python
df1=pd.DataFrame(np.random.rand(2,2),index=[1,2],columns=['a','b'])
df2=pd.DataFrame(np.random.rand(2,2),index=[1,2],columns=['a','b'])
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])
```

```python
df3=pd.DataFrame(np.random.rand(2,2),index=[1,2],columns=['a','b'])
df4=pd.DataFrame(np.random.rand(4,5),index=[1,2,3,4],columns=['a','b','c','d','e'])
```

```python
df3
```

```python
df4
```

```python
pd.concat([df3,df4])
```

```python
pd.concat([df3,df4],ignore_index=True)
```

```python
pd.concat([df3,df4],join='inner')
```

```python
df1 = pd.DataFrame({'student':['Ram','Vinay','Raju','Ravi'],
                    'group':['ECE','EEE','EEE','CSE']})

df2 = pd.DataFrame({'student':['Raju','Ram','Vinay','Ravi'],
                    'marks':[40,50,60,89]})
```

```python
pd.concat([df1,df2])
```

```python
pd.merge(df1,df2)
```
